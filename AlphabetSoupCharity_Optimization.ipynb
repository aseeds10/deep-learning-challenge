```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
## Preprocessing
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df.drop(columns=["EIN", "NAME"], inplace=True)
application_df.head()
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"
applic_type_value_counts = application_df["APPLICATION_TYPE"].value_counts()
applic_type_value_counts
```

```python
# CHANGE CUT-OFF POINT FOR OTHER APPLIC TYPE
#empty list to set up conditional to fill with applic types that pass threshold
application_types_to_replace = []

# Iterate over the counts
for each, count in applic_type_value_counts.items():
    if count < 700:
        application_types_to_replace.append(each)

# print(application_types_to_replace)

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure replacement was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts to identify and replace with "Other"
classification_value_counts = application_df["CLASSIFICATION"].value_counts()

for classification, count in classification_value_counts.items():
    if count > 1:
        print(f"{classification}: {count}")
```

```python
#KEEP SAME CUT-OFF FOR CLASSIFICATION
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = []

# Iterate over the counts
for classification, count in classification_value_counts.items():
    if count < 1000:
        classifications_to_replace.append(classification)

print(classifications_to_replace)

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure replacement was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
#check counts of SPECIAL_CONSIDERATIONS to see if noisy data
application_df["SPECIAL_CONSIDERATIONS"].value_counts()
#NOTE: vast majority are N so will drop to quiet noise in model
```

```python
#check counts of STATUS to see if noisy data
application_df["STATUS"].value_counts()
#NOTE: vast majority are 1 so will drop to quiet noise in model
```

```python
# Drop the non-beneficial columns, 'SPECIAL_CONSIDERATIONS' and 'STATUS'.
application_df.drop(columns=["SPECIAL_CONSIDERATIONS", "STATUS"], inplace=True)
application_df.head()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
# application_df_copy =  application_df.copy()
# application_dummies = pd.get_dummies(application_df, \
#                                      columns=['APPLICATION_TYPE', \
#                                               'CLASSIFICATION', 'AFFILIATION', \
#                                               'USE_CASE', 'ORGANIZATION', \
#                                               'INCOME_AMT', 'ASK_AMT'], \
#                                      drop_first=True)

application_dummies = pd.get_dummies(application_df, \
                                     columns=['APPLICATION_TYPE', \
                                              'CLASSIFICATION', 'AFFILIATION', \
                                              'USE_CASE', 'ORGANIZATION', \
                                              'INCOME_AMT'], \
                                     drop_first=True)

# print(dummies[:3])
application_dummies.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = application_dummies["IS_SUCCESSFUL"]
X = application_dummies.drop(["IS_SUCCESSFUL"], axis=1)

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
X_train_scaled[:1]
```

```python
## Compile, Train and Evaluate the Model
```

```python
print(len(X_train.columns))
```

```python
# MODEL 1
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train.columns)
hidden_nodes_layer1 =  10
hidden_nodes_layer2 = 5

nn = tf.keras.models.Sequential()

# Input layer
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="tanh")
)

# 1st hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="tanh"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled,y_train,epochs=100)
```

```python
# Evaluate the model using the test data - MODEL 1
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
